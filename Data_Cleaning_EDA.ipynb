```python
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
sns.countplot(x = 'gender',data = df)
plt.title ("No of patients by gender")
```

```python
medical_condition_vs_gender = pd.crosstab(df['gender'],df['medical_condition'])

medical_condition_vs_gender.plot(kind = 'bar')
plt.title("Medical Condition Distribution by Gender")
plt.ylabel("Count")
```

```python
blood_count = df['blood_type'].value_counts()
blood_count
```

```python
blood_count.plot(kind ='pie')
plt.title("Distribution of Blood Types")
plt.legend(title = ('blood_type'),loc = 'upper right')
```

```python
med_condition_ct = pd.crosstab(df['medical_condition'], df['medication'])

plt.figure(figsize=(10, 6))
sns.heatmap(med_condition_ct, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Frequency Heatmap: Medication vs. Medical Condition')
plt.show()
```
